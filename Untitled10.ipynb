## Обучение модели

удалим пока пропущенные значения, хотя во временном ряде так делать нельзя

```python
train = train.dropna()
```

Раздели обучающую выборку на features и target

```python
features_train = train.drop(col_name, axis=1)
target_train = train[col_name]
```

```python
grid_rdf = {'n_estimators': range(5, 50, 10), 'max_depth': range(3, 15, 5)}
```

- Модель линейной регрессии

```python
model = LinearRegression()
grid = {}
gsearch = GridSearchCV(model, grid, cv=TimeSeriesSplit(n_splits=5).split(features_train), scoring="neg_root_mean_squared_error")
gsearch.fit(features_train, target_train)
print(gsearch.best_params_)
print(gsearch.best_score_*(-1))
```

- Модель случайного леса

```python
model = RandomForestRegressor(random_state=RANDOM_STATE)
grs = GridSearchCV(model, grid_rdf, cv=TimeSeriesSplit(n_splits=5).split(features_train), scoring="neg_root_mean_squared_error")
grs.fit(features_train, target_train)
print(grs.best_params_)
print(grs.best_score_*(-1))
```

```python
# Import the library
import pmdarima as pm
```

```python

model = pm.auto_arima(train, 
                      m=12, 
                      seasonal=True,
                      start_p=0, start_q=0, max_order=4, test='adf',error_action='ignore',  
                      suppress_warnings=True,
                      stepwise=True, 
                      trace=True)

```

```python
model.summary()
```

```python
forecast=model.predict(n_periods=12, 
                       return_conf_int=True)
```

```python
forecast_df = pd.DataFrame(forecast[0],
                           index = forecast[0].index,
                           columns=['Prediction'])
```
